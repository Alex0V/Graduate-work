```python
import random

# Кількість користувачів і кількість курсів
num_users = 40
num_courses = 30

# Зберігатимемо оцінки в список кортежів (user_id, course_id, score)
ratings = []

# Створюємо словник для зберігання пройдених курсів та їх оцінок кожним користувачем
user_ratings = {}

# Генеруємо випадкові оцінки для кожного користувача
for user_id in range(2, num_users + 1):
    # Випадково визначаємо кількість курсів, які оцінює користувач
    k = random.randint(1, num_courses-7)
    
    # Випадково обираємо k курсів (без повторень)
    chosen_courses = random.sample(range(1, num_courses + 1), k)
    
    # Зберігаємо унікальні id курсів та їх оцінки для поточного користувача
    user_ratings[user_id] = []
    for course_id in chosen_courses:
        score = random.randint(1, 5)
        user_ratings[user_id].append((course_id, score))

```

```python
count = 0
# Виведемо унікальні id курсів та їх оцінки для кожного користувача
for user_id, courses_ratings in user_ratings.items():
    for course_id, score in courses_ratings:
        count+=1
print(count)
```

```python
count = 0
# Виведемо унікальні id курсів та їх оцінки для кожного користувача
for user_id, courses_ratings in user_ratings.items():
    if len(courses_ratings)>19:
        print(f"Користувач {user_id}: ({len(courses_ratings)})")
        count+=1

print(count)
```
